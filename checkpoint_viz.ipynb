```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

months = [
    ("Feb (Winter)", "images/sequoia_feb_9_2025.png"),
    ("Apr (Spring)", "images/sequoia_apr9_2025.png"),
    ("Jun (Summer)", "images/sequoia_jun9_2025.png"),
    ("Aug (Summer)", "images/sequoia_aug9_2025.png"),
    ("Oct (Fall)", "images/sequoia_oct9_2025.png"),
    ("Dec (Winter)", "images/sequoia_dec9_2025.png"),
]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for ax, (month, path) in zip(axes.flatten(), months):

    img = mpimg.imread(path)

    ax.imshow(img)
    ax.set_title(month)
    ax.axis("off")

fig.suptitle(
    "Seasonal Vegetation Changes in Sequoia National Park (2025)",
    fontsize=18
)


plt.savefig("seasonal_comparison.png", dpi=300)
```
